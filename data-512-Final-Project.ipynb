# Introduction 
 
OASDI stands for Old-Age, Survivors, Disability Insurance. It is part of Social Security Disability program. It is provided as support to people who are retired and families (dependent spouses and children) of workers have died. Additional details on its history can be found here on Wikipedia - https://en.wikipedia.org/wiki/Social_Security_(United_States)  
  
SSI stands for Supplemental Security Income. It is part of Social Security and involves monthly payment as support to people with limited income and people who are disabled, blind or aged 65 years and above. Additional details about SSI can be found here - https://www.ssa.gov/ssi/text-over-ussi.htm   
  
Both programs are administered by Social Security Administration (SSA) - https://www.ssa.gov/.
The key difference between these two programs, as mentioned in the link (https://www.ssa.gov/ssi/text-over-ussi.htm) is:
“Social Security benefits may be paid to you and certain members of your family if you are “insured” meaning you worked long enough and paid Social Security taxes. Unlike Social Security benefits, SSI benefits are not based on your prior work or a family member's prior work.”  
  
Each year congressional reports are published with information of SSI and OSDI payments made by each state. This report includes information about no. of recipients in each category and the total payments made for each state.  
  
I have analyzed this data for the years 2013, 2014, 2015, 2016 and 2017. I performed this analysis as part of my final project for Human-Centered Data Science course (DATA 512) during Fall 2018 at the University of Washington. Keeping in mind the human centered aspect of this course, my intent was to get a sense of distribution of people benefiting from OASDI and SSI across US. 
 
OASDI and SSI are very important support systems US has for those who are in need of financial assistance for survival and critical health needs. So it is important to analyze the spending to ensure there is equitable distribution and there are no anomalies. It is also important to analyze and identify if there are any strange trends which may give rise some to more research questions.

```python
# All source filenames
data_workbooks_files = ['Data/cong_stats13.xlsx', \
               'Data/cong_stats14.xlsx', \
               'Data/cong_stats15.xlsx', \
               'Data/cong_stats16.xlsx', \
               'Data/cong_stats17.xlsx']

# Data Periods in years
data_periods = ['2013', '2014', '2015', '2016', '2017']

# Data columns for OASDI
oasdi_df_columns = ['period', \
                    'state', \
                    'total_ben', \
                    'retired_workers', \
                    'disabled_workers', \
                    'widowers_parents', \
                    'spouses', \
                    'children', \
                    'age_65_older', \
                    'amt_total_ben', \
                    'amt_retired_workers', \
                    'amt_widowers_parents']

# Cell information (row, col) of each data column mentioned in data columns above. 
# Each pair below corresponds to the row no. and col no. of corresponding cell in the excel sheet for OASDI where the data
# for the corresponding column is stored.
oasdi_column_cells = [(5,3), (5,4), (5,5), (5,6), (5,7), (5,8), (5,12), (5,9), (5,10), (5,11)]

# States to be exclued in OASDI
oasdi_exclude_states = ['AS', 'GU', 'MP', 'PR', 'VI']

# Data columns for SSI
ssi_df_columns = ['period', \
                  'state', \
                  'total_ben', \
                  'aged', \
                  'blind', \
                  'disabled', \
                  'with_oasdi', \
                  'aged_65_older', \
                  'amt_total_ben', \
                  'amt_aged', \
                  'amt_blind', \
                  'amt_disabled']

# Cell information (row, col) of each data column mentioned in data columns above. 
# Each pair below corresponds to the row no. and col no. of corresponding cell in the excel sheet for SSI where the data
# for the corresponding column is stored.
# It appears that the format slightly changed for SSI for years 2016 and 2017. So, using two different variables to 
# fetch accordingly
ssi_column_cells_2013_to_2015 = [(5,3), (5,4), (5,5), (5,6), (5,11), (5,12), (5,7), (5,8), (5,9), (5,10)]
ssi_column_cells_2016_2017 = [(6,3), (6,4), (6,5), (6,6), (6,9), (6,14), (6,10), (6,11), (6,12), (6,13)]

# States to be exclued in SSI
ssi_exclude_states = ['MP']

# Population data in exact numbers for the years 2013 through 2017 in order.
population_data = [316234505, 318622525, 321039839, 323405935, 325719178]
```

## Data Extraction
  
The following function extracts the data from the input excel files using the setup data defined above.  
It loops through all the files, processing one file a time.  
For each file, it loops through all the sheets.  
If the sheet's name has 'OASDI' in the name, it uses the setup data corresponding to OASDI.  
If the sheet's name has 'SSI' in the name, it uses the setup data corresponding to SSI.  
For SSI, based on the year the sheet corresponds to, the appropriate setup data is used.
xlrd library is used to open the excel files and read particular cells.  
Once all the files are processed, the function returns two dataframes - one each for OASDI and SSI.

```python
# Function to extract the data from the input excel files using the setup data defined above

def extract_data():
    
    df_oasdi = {key:[] for key in oasdi_df_columns}
    df_ssi = {key:[] for key in ssi_df_columns}
    
    j = 0
    for workbook_name in data_workbooks_files:
        workbook = xlrd.open_workbook(workbook_name)
        period = data_periods[j]
        j += 1
        for sheet_name in workbook.sheet_names():
            if 'OASDI' in sheet_name:
                state = sheet_name[:2]
                if state not in oasdi_exclude_states:
                    df_oasdi['period'].append(period)
                    df_oasdi['state'].append(state)
                    worksheet = workbook.sheet_by_name(sheet_name)
                    for i in range(2, len(oasdi_df_columns)):
                        row, col = oasdi_column_cells[i-2]
                        df_oasdi[oasdi_df_columns[i]].append(worksheet.cell_value(row-1, col-1))
            if 'SSI' in sheet_name:
                state = sheet_name[:2]
                if state not in ssi_exclude_states:
                    df_ssi['period'].append(period)
                    df_ssi['state'].append(state)
                    worksheet = workbook.sheet_by_name(sheet_name)
                    for i in range(2, len(ssi_df_columns)):
                        if period in ['2013', '2014', '2015']:
                            row, col = ssi_column_cells_2013_to_2015[i-2]
                        else:
                            row, col = ssi_column_cells_2016_2017[i-2]
                        df_ssi[ssi_df_columns[i]].append(worksheet.cell_value(row-1, col-1))
        print(workbook_name + '..done!')
    return pd.DataFrame(df_oasdi), pd.DataFrame(df_ssi)
```

The following code calls the function defined above to create the dataframes for OASDI and SSI.  
It also adds two new columns to the dataframes to store the spending per person. It is computed by dividing the total amount spent by total no. of beneficiaries.

```python
df_oasdi, df_ssi = extract_data()
df_oasdi['amt_per_ben'] = df_oasdi['amt_total_ben'] / df_oasdi['total_ben']
df_ssi['amt_per_ben'] = df_ssi['amt_total_ben'] / df_ssi['total_ben']
```
